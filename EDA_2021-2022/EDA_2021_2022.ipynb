```python
dd_2021=df_2021['ended_at']-df_2021['started_at']
dd_2021=pd.DataFrame({'Date':dd_2021})
dd_2021['Date'].dt.components
```

Are there any common stations in years 2021 and 2022?

```python
l1=df_2021_longest_used['start_station_name']
l2=df_2022_longest_used['start_station_name']
common_stations= list(set(l1).intersection(l2))

d_common_stations=[]
for i in common_stations:
    d_common_stations.append(df_2021_longest_used[df_2021_longest_used['start_station_name']==i])
    d_common_stations.append(df_2022_longest_used[df_2022_longest_used['start_station_name']==i])

temp_df = pd.DataFrame() #Temporary empty dataframe
for i in range(len(d_common_stations)):
    df_1=pd.DataFrame(d_common_stations[i])
    temp_df = temp_df.append(df_1,ignore_index=True) #Moving the contents of newly created dataframe to the temporary dataframe

temp_df['year']=np.where((((temp_df.index ==0 )) | ((temp_df.index < len(d_common_stations)-1) & (temp_df.index % 2 ==0))), 2021,2022)
```

```python
fig=plt.figure(figsize=(12,6))
sns.barplot(data=temp_df, x='start_station_name',y='Duration', hue='rideable_type')
plt.xticks(rotation=90)
```
